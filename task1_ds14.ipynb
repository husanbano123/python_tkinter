```python
import math
import tkinter as tk
from tkinter import ttk
```

```python


def calculate_side():
    choice = choice_var.get()

    if choice == "Right-angle triangle":
        right_angle_triangle()
    elif choice == "Side opposite to 30 degrees":
        thirty_degree_triangle()
    elif choice == "Side opposite to 45 degrees":
        forty_five_triangle()
    elif choice == "Side opposite to 60 degrees":
        sixty_triangle()
    else:
        result_label.config(text="Invalid choice! Please enter a number between 1 and 4.")

def right_angle_triangle():
    side1 = float(side1_entry.get())
    side2 = float(side2_entry.get())
    hypt = math.sqrt(side1**2 + side2**2)
    result_label.config(text=f"The hypotenuse is: {round(hypt, 2)} cm")

def thirty_degree_triangle():
    hypt = float(hypt_entry.get())
    side_30 = 1/2 * hypt
    result_label.config(text=f"The side opposite to 30 degrees is {round(side_30, 2)} cm")

def forty_five_triangle():
    hypt = float(hypt_entry.get())
    side_45 = 1/(math.sqrt(2)) * hypt
    result_label.config(text=f"The side opposite to 45 degrees is {round(side_45, 2)} cm")

def sixty_triangle():
    hypt = float(hypt_entry.get())
    side_60 = (math.sqrt(3))/2 * hypt
    result_label.config(text=f"The side opposite to 60 degrees is {round(side_60, 2)} cm")

# Create the main window
root = tk.Tk()
root.title("Triangle Calculator")

# Create and configure widgets
frame = ttk.Frame(root, padding="10")
frame.grid(row=0, column=0, sticky=(tk.W, tk.E, tk.N, tk.S))

choice_var = tk.StringVar()
choice_label = ttk.Label(frame, text="Choose a triangle:")
choice_label.grid(row=0, column=0, columnspan=2, pady=(0, 10))

choices = ["Right-angle triangle", "Side opposite to 30 degrees", "Side opposite to 45 degrees", "Side opposite to 60 degrees"]
choice_menu = ttk.Combobox(frame, values=choices, textvariable=choice_var)
choice_menu.grid(row=1, column=0, columnspan=2, pady=(0, 10))

side1_label = ttk.Label(frame, text="Side 1:")
side1_label.grid(row=2, column=0, pady=(0, 10))
side1_entry = ttk.Entry(frame)
side1_entry.grid(row=2, column=1, pady=(0, 10))

side2_label = ttk.Label(frame, text="Side 2:")
side2_label.grid(row=3, column=0, pady=(0, 10))
side2_entry = ttk.Entry(frame)
side2_entry.grid(row=3, column=1, pady=(0, 10))

hypt_label = ttk.Label(frame, text="Hypotenuse:")
hypt_label.grid(row=4, column=0, pady=(0, 10))
hypt_entry = ttk.Entry(frame)
hypt_entry.grid(row=4, column=1, pady=(0, 10))

calculate_button = ttk.Button(frame, text="Calculate", command=calculate_side)
calculate_button.grid(row=5, column=0, columnspan=2, pady=(10, 0))

result_label = ttk.Label(frame, text="")
result_label.grid(row=6, column=0, columnspan=2, pady=(10, 0))

# Start the main event loop
root.mainloop()
```
